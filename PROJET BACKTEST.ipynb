```python
import pandas as pd 
```

```python
df=pd.read_csv('ETC-USDT.csv')
df
```

```python
df["date"]= pd.to_datetime(df["date"], unit="ms")
df
```

```python
df = df.set_index(df["date"])
del df["date"]
df
```

```python
df["Ma 100"]=df["close"].rolling(100).mean()
df
```

```python
df["MA 200"]=df["close"].rolling(200).mean()
df
```

```python
df["buy signal"]=False
df["sell signal"]=False
df.loc[df["Ma 100"] > df["MA 200"], "buy signal"]= True
df.loc[df["Ma 100"] < df["MA 200"], "sell signal"]= True
df
```

```python
df["sell signal"].value_counts()
```

```python
df["buy signal"].value_counts()
```

```python
balance=100000
position=None
for index, row in df.iterrows():
    if position is None and row["buy signal"] is True:
        buy_price=row["close"]
        position_seize=1000 
        position={
            "buy_price":buy_price,
            "position_seize":position_seize,
            
        }
        print(f"{index}-buy for{position_seize}$ price of {buy_price}$")
        
    elif position is not None and row ["sell signal"] is True:
        sell_price=row["close"]
        trade_result=(sell_price-position["buy_price"])/ position["buy_price"]
        balance=balance+trade_result*position["position_seize"]
        print(f"{index} -sell for {balance}$ at the price of {sell_price}$ ({trade_result*100}%)")
        position = None
        
         
        
        
                       
                                       
     
```
